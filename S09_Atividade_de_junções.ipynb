# Crie um dataframe único com a informação do GDP e quantidade de matrículas por nível de educação por ano.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
educ = 'https://raw.githubusercontent.com/abnr/ml-data/main/SYB61_T07_Education.csv'
gdp = 'https://raw.githubusercontent.com/abnr/ml-data/main/India_World_Bank_Info.csv'
```

```python
df_educ = pd.read_csv(educ, header=1,   sep = '\t', encoding='unicode_escape')
df_educ.head()
```

```python
df_gdp = pd.read_csv(gdp, sep='\t', on_bad_lines='error', skiprows=4)
df_gdp.head()
```

```python
# Somente a educação, ano e valores:
df_educ_formated = df_educ[["Year", "Series", "Value"]]
df_educ_formated.head()
```

```python
# Filtrando e reorganizando o DataFrame
df_just_gdp = df_gdp.loc[df_gdp["Indicator Name"] == "GDP (current US$)"].melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="GDP")

# Redefinindo o índice e removendo a primeira linha:
df_just_gdp = df_just_gdp.iloc[1:].reset_index(drop=True)
df_just_gdp.head()
```

```python
# Conversão para o ano ficar em inteiro:
df_just_gdp["Year"] = df_just_gdp["Year"].astype('int64')
```

```python
df_just_gdp.dtypes
```

# Joins:

```python
gdp_and_educ = pd.merge(left=df_educ_formated, right=df_just_gdp, on=["Year"], how="inner")
gdp_and_educ.head()
```

# O aumento das matrículas na educação primária, secundária e terciária acompanhou o aumento do GDP per carpita nos últimos 15 anos na Índia?

```python
# Filtrar dados da Índia
india_data = gdp_and_educ[gdp_and_educ['Country Name'] == 'India']
```

```python
# Agrupar dados por ano e calcular a média das matrículas e dpo GDP per carpita
education_gdp = india_data.groupby('Year').agg({
    'Value': 'sum',
    'GDP': 'mean'
}).reset_index()
```

```python
# Filtrando dados para gráficos de linhas e limitando aos anos desejados
education_gdp = education_gdp[(education_gdp['Year'] >= 2009) & (education_gdp['Year'] <= 2023)]
```

```python
# Filtrando dados para gráficos de linhas
years = education_gdp['Year']
total_students = education_gdp['Value']
gdp = education_gdp['GDP']
```

```python
# Plotando gráfico
fig, ax1 = plt.subplots(figsize=(12, 8))

# Gráfico de linha para total de estudantes
ax1.plot(years, total_students, color='blue', marker='o', linestyle='-', label='Total Students')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Students', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Eixo y2 para GDP per capita
ax2 = ax1.twinx()
ax2.plot(years, gdp, color='green', marker='o', linestyle='-', label='GDP per Capita (current US$)')
ax2.set_ylabel('GDP per Capita (current US$)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Título do gráfico
plt.title('Total Students vs GDP per Capita in Índia (2009-2023)')

# Adicionando legendas
fig.legend(loc='upper right', bbox_to_anchor=(0.85, 0.85))

# Exibir o gráfico
plt.show()
```
