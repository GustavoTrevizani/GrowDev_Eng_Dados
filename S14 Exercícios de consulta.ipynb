2 - Encontre o nome e o endereço dos clientes que fizeram um pedido em 2024.

```python
query2 = """
SELECT c.nome, c.endereco
FROM Clientes c
JOIN Pedidos p ON c.id_cliente = p.id_cliente
WHERE EXTRACT(YEAR FROM p.data_pedido) = 2024
"""
pd.read_sql(query2,conn)
```

3 - Liste os produtos cujo preço está entre 50 e 100.

```python
query3 = """
SELECT * 
FROM produtos
WHERE preco BETWEEN 50 AND 100;
"""
pd.read_sql(query3,conn)
```

4 - Conte o número total de pedidos feitos no mês de julho de 2024.

```python
query4 = """
SELECT COUNT(*) FROM Pedidos p
WHERE EXTRACT(MONTH FROM p.data_pedido) = 7
AND EXTRACT(YEAR FROM p.data_pedido) = 2024;
"""
pd.read_sql(query4,conn)
```

5 - Encontre o total pago por cada pedido.

```python
query5 = """ 
SELECT p.id_pedido, SUM(pg.valor_pago)
FROM pedidos p
JOIN pagamentos pg ON p.id_pedido = pg.id_pedido
GROUP BY p.id_pedido;
"""
pd.read_sql(query5,conn)
```

6 - Liste o nome dos clientes e o valor total dos seus pedidos.

```python
query6 = """
SELECT c.nome, SUM(p.valor_total)
FROM clientes c
JOIN pedidos p ON c.id_cliente = p.id_cliente
GROUP BY c.nome;
"""
pd.read_sql(query6,conn)
```

7 - Encontre os produtos com estoque menor que 10

```python
query7 = """ 
SELECT *
FROM produtos
WHERE estoque < 10;
"""
pd.read_sql(query7,conn)
```

8 - Liste os pedidos com seu status e a quantidade de itens em cada pedido.

```python
query8 = """ 
SELECT p.id_pedido, p.status, SUM(ip.quantidade)
FROM pedidos p
JOIN itens_pedido ip ON p.id_pedido = ip.id_pedido
GROUP BY p.id_pedido, p.status;
"""
pd.read_sql(query8,conn)
```

9 - Encontre os produtos que não foram vendidos em nenhum pedido.

```python
query9 = """ 
SELECT *
FROM produtos p
WHERE p.id_produto NOT IN (
    SELECT id_produto
    FROM itens_pedido
);
"""
pd.read_sql(query9,conn)
```

10 - Liste o nome dos 10 maiores clientes e a quantidade de pedidos feitos por cada um. (por quantidade de pedidos)

```python
query10 = """
SELECT c.nome, COUNT(p.id_pedido) AS quantidade_pedidos
FROM clientes c
JOIN pedidos p ON c.id_cliente = p.id_cliente
GROUP BY c.nome
ORDER BY quantidade_pedidos DESC
LIMIT 10;
"""
pd.read_sql(query10,conn)
```

11 - Encontre o produto mais caro e o mais barato.

```python
#Produto mais caro
query11 = """ 
SELECT *
FROM produtos p
ORDER BY p.preco DESC
LIMIT 1;
"""
pd.read_sql(query11,conn)
```

```python
#Produto mais barato
query11 = """ 
SELECT *
FROM produtos p
ORDER BY p.preco ASC
LIMIT 1;
"""
pd.read_sql(query11,conn)
```

12 - Calcule a idade média dos clientes.

```python
query12 = """ 
SELECT AVG(EXTRACT(YEAR FROM c.data_nascimento))
FROM clientes c;
"""
pd.read_sql(query12,conn)
```

13 - Liste os clientes que possuem o mesmo sobrenome.


```python
query13 = """
WITH Sobrenomes AS (
    SELECT 
        TRIM(SPLIT_PART(nome, ' ', ARRAY_LENGTH(STRING_TO_ARRAY(nome, ' '), 1))) AS sobrenome,
        id_cliente,
        nome
    FROM Clientes
),
Duplicados AS (
    SELECT sobrenome
    FROM Sobrenomes
    GROUP BY sobrenome
    HAVING COUNT(*) > 1
)
SELECT s.id_cliente, s.nome, s.sobrenome
FROM Sobrenomes s
JOIN Duplicados d ON s.sobrenome = d.sobrenome
ORDER BY s.sobrenome, s.id_cliente;
"""
pd.read_sql(query13,conn)
```

14 - Encontre os pedidos feitos por clientes cujo nome começa com a letra 'A'.


```python
query14 = """ 
SELECT *
FROM pedidos p
JOIN clientes c ON p.id_cliente = c.id_cliente
WHERE c.nome LIKE 'A%';
"""
pd.read_sql(query14,conn)
```

15 - Liste os produtos, sua categoria e uma indicação se estão em estoque ou esgotados (usando CASE).

```python
query15 = """
SELECT p.id_produto, p.nome, p.categoria,
CASE
    WHEN p.estoque > 0 THEN 'Em estoque'
    ELSE 'Esgotado'
END
FROM produtos p;
"""
pd.read_sql(query15,conn)
```
